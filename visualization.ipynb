```python
# Import necessary libraries
import plotly.graph_objects as go
import datetime
from dateutil.relativedelta import relativedelta
from calc import amortization_schedule, adjusted_amortization_schedule
import plotly.io as pio
pio.renderers.default = "notebook"  # or "jupyterlab" if you are using JupyterLab
```

```python
# Define loan parameters
loan_value = 300000
interest = 9.5
term = 30
max_monthly_payment = 2500
down_payment = 150000
start_date = datetime.datetime.today()
property_taxes = 7000
home_insurance = 2000  # Changed 'insurance' to 'home_insurance' to match the function argument
hoa_fees = 0  # Changed 'hoa' to 'hoa_fees' to match the function argument
pmi = 0
```

```python
# Get the adjusted amortization schedule
schedule = adjusted_amortization_schedule(
    loan_value, interest, term, max_monthly_payment,
    down_payment, start_date, property_taxes,
    home_insurance, hoa_fees, pmi
)
```

```python
# Extract data for plotting
months = [payment['Month'] for payment in schedule]
monthly_payments = [payment['Monthly Payment'] for payment in schedule]
total_payments_sum = [payment['Total Payments Sum'] for payment in schedule]
interest_sum = [payment['Interest Sum'] for payment in schedule]
principal_sum = [payment['Principal Sum'] for payment in schedule]
property_taxes_sum = [payment['Property Taxes Sum'] for payment in schedule]
home_insurance_sum = [payment['Home Insurance Sum'] for payment in schedule]
hoa_fees_sum = [payment['HOA Fees Sum'] for payment in schedule]
pmi_sum = [payment['PMI Sum'] for payment in schedule]
```

```python
# Create the plot
fig = go.Figure()
```

```python
# Add traces for each line
fig.add_trace(go.Scatter(x=months, y=monthly_payments, mode='lines', name='Monthly Payments'))
fig.add_trace(go.Scatter(x=months, y=total_payments_sum, mode='lines', name='Total Payments Sum'))
fig.add_trace(go.Scatter(x=months, y=interest_sum, mode='lines', name='Interest Sum'))
fig.add_trace(go.Scatter(x=months, y=principal_sum, mode='lines', name='Principal Sum'))
fig.add_trace(go.Scatter(x=months, y=property_taxes_sum, mode='lines', name='Property Taxes Sum'))
fig.add_trace(go.Scatter(x=months, y=home_insurance_sum, mode='lines', name='Home Insurance Sum'))
fig.add_trace(go.Scatter(x=months, y=hoa_fees_sum, mode='lines', name='HOA Fees Sum'))
fig.add_trace(go.Scatter(x=months, y=pmi_sum, mode='lines', name='PMI Sum'))
```

```python
# Set the title and labels
fig.update_layout(
    title='Amortization Schedule Over Time',
    xaxis_title='Month',
    yaxis_title='Amount ($)',
    template='plotly_dark'
)

# Show the plot
fig.show()
```
